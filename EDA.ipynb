```python
# importar librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# cargar Dataset 
car_data = pd.read_csv('vehicles_us (1).csv')
display(car_data)
```

```python
# información general de los datos
car_data.info()
```

```python
# convertir los datos al tipo correspondiente 
car_data['model_year'] = car_data['model_year'].astype('Int64')
car_data['date_posted'] = pd.to_datetime(car_data['date_posted'], format='%Y-%m-%d')
display(car_data)
```

```python
# Datos con el tipo de dato correspondiente.
car_data.info()
```

```python
# Encontrar la distrinución de los valores ausentes. 
car_data.isna().sum()/len(car_data)*100
```

```python
# observar valores únicos 
car_data['model_year'].unique()
```

```python
# observar valores únicos 
car_data['cylinders'].unique()
```

```python
# Reemplazar los valores ausentes por 0
car_data['model_year'] = car_data['model_year'].fillna(0) 
car_data['cylinders'] = car_data['cylinders'].fillna(0) 
car_data.isna().sum()
```

```python
# correlación de prcio segun el modelo 
model_per_price= car_data.groupby('model_year')['price'].sum().sort_values().reset_index()
model_per_price['model_year'].corr(model_per_price['price'])*100
model_per_price.plot(kind= 'hist',
                     ylabel= 'PRECIO',
                     xlabel='MEDELO DEL AÑO',
                     title= 'PRECIO SEGUN EL AÑO DEL VEHICULO')
```
